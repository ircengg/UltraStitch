```python
import h5py
import shutil
import os

import tempfile

def compact_h5_safe(original_path):
    dir_ = os.path.dirname(original_path)

    with tempfile.NamedTemporaryFile(
        suffix=".h5",
        dir=dir_,
        delete=False
    ) as tmp:
        tmp_path = tmp.name

    try:
        with h5py.File(original_path, "r") as src, \
             h5py.File(tmp_path, "w") as dst:

            for name in src:
                src.copy(name, dst)

            for k, v in src.attrs.items():
                dst.attrs[k] = v

        os.replace(tmp_path, original_path)

    except Exception:
        os.remove(tmp_path)
        raise


compact_h5_safe(r"C:\Users\Yogesh Chaudhary\Desktop\ScanReg\IOCL Silcher AUT\IOCL Silcher-AUT.h5")
```
